#Depression and Experience of violence

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data preparation (replace this with your actual data loading)
# data = pd.read_csv('your_dataset.csv')

# Set the order for depression levels
depression_order = ["Minimal", "Mild", "Moderate", "Severely Moderate", "Severe Depression"]

# Set 'Depression_level' as a categorical variable with ordered categories
data['Depression_level'] = pd.Categorical(data['Depression_level'], categories=depression_order, ordered=True)

# Calculate the total count for each experience of violence
violence_counts = data['Experience of violence'].value_counts(normalize=True) * 100

# Calculate the percentage of each depression level within each type of violence
depression_by_violence = data.groupby(['Experience of violence', 'Depression_level']).size().unstack(fill_value=0)
depression_by_violence_percentage = depression_by_violence.div(depression_by_violence.sum(axis=1), axis=0) * 100

# Plotting
fig, ax = plt.subplots(figsize=(8, 4))

# Create stacked bar chart with specified order
depression_by_violence_percentage[depression_order].plot(kind='bar', stacked=True, ax=ax, color=['#a1c4fd', '#4facfe', '#00c6ff', '#0072bb', '#001f3f'])

# Add annotations for total percentages of each experience type
for i, (violence_type, pct) in enumerate(violence_counts.items()):
    ax.text(i, 105, f'{pct:.1f}%', ha='center', va='bottom', color='black', fontsize=10)

# Set plot titles and labels
ax.set_title("Distribution of Depression Levels by Experience of Violence\n")
ax.set_xlabel("Experience of Violence")
ax.set_ylabel("Percentage of Depression Levels")

# Customize x-ticks to show only unique labels without duplication
ax.set_xticklabels(depression_by_violence_percentage.index, rotation=0)

# Customize legend
plt.legend(title='Depression Level', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
